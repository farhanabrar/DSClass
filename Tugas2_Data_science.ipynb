```python
 q1 = df.quantile(0.25)
 q3 = df.quantile(0.75)
 iqr = q3 -q1
 iqr
```

```python
import warnings
warnings.filterwarnings('ignore')
df_align, iqr_new = df.align(iqr, axis = 1, copy=False, join="outer")
outlier_filter = (df < q1 - 1.5 * iqr_new) | (df > q3 + 0.75 * iqr_new)
outlier_filter
```

```python
df[outlier_filter['No. of Movies']]\
.loc[:, ['Country','No. of Movies']]\
.sort_values(by=['No. of Movies'], ascending=False)
```
