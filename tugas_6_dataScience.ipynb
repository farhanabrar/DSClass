tugas berikutnya adalah melakukan imputasi. yaitu mengisi data yang kosong dengan sebuah nilai. beberapa teknik imputasi yang akan dilakukan pada tugas ini adalah: MEAN, ARBITRARY, END OF TAIL, MODUS (NILAI TERBANYAK), KONSTANTA

```python
path = '/content/netflix price in different_edited.csv'
pd.set_option('display.max_rows', 500)
data = pd.read_csv(path)
data
```

imputasi MEAN, yaitu mengisi data yang hilang (NaN) dengan nilai rata-rata dari kolom tersebut

```python
data.fillna(data.mean())
```

Imputasi ARBITRARY yaitu mengisi nilai kosong (NaN) dengan nilai suka-suka 

```python
data.fillna(3000)
```

```python
test = imputer.transform(data)
test
```

```python
category = {'cat' : ['Furniture', 'Furniture', 'Office Supplies', 'Technology', 'Furniture', 'Technology', np.nan, 'Furniture', 'Office Supplies', np.nan, 'Furniture']}
data2 = pd.DataFrame(category)
data2
```

untuk metode imputasi menggunakan most frequent atau modus, mengganti index ke 6 dan 10 dengan 'furniture' karena category furniture merupakan modus dari data atau value yang paling sering keluar

```python
from feature_engine.imputation import RandomSampleImputer
data3 = {'category' : ['Furniture', 'Furniture', 'Office Supplies', 'Technology', 'Furniture', 'Technology', np.nan, 'Furniture', 'Office Supplies', np.nan, 'Furniture'],
         'sold': [np.nan, '150', '500', '100', np.nan, '300', '320', '490', '200', '120', '420']}
df =pd.DataFrame(data3)
df
```

```python
imputer = RandomSampleImputer(random_state = 29)

#Fit imputer ke set 
imputer.fit(df)

#Mengubah data
test_trans = imputer.transform(df)
test_trans
```
